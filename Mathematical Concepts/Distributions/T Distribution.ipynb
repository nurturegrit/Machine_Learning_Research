## Introduction

The t-distribution is a kind of probability distribution that is symmetric and bell shaped, similar to Normal Distribution but with heaviour tails.
The Exact Shape of t-distribution depends on `degree of freedom (df)`.

The t-distribution, also known as Student's t-distribution, is a type of probability distribution that is symmetric and bell-shaped, like the normal distribution, but has heavier tails. This means it is `more prone to producing values that fall far from its mean`.

#### Now asking a question. If we have a Normal Distribution (which is naturally occuring), why do we need t-distribution?

Ans : The t-distribution is used when we are estimating the mean of a normally distributed population in situations where the sample size is small and the population standard deviation is unknown. It allows for a more accurate estimate by accounting for the `extra variability/uncertainity` that comes with smaller samples. This `uncertainity` is applied through `degree of freedom`.

## Understanding the Extra Variability

   * Sample Variability: When you have a small sample, the sample mean and sample standard deviation can vary significantly from sample to sample. This variability is higher compared to larger samples because each individual data point has a greater influence on the overall statistics.

   * Estimating Population Standard Deviation: With small samples, our estimate of the population standard deviation (using the sample standard deviation) is less precise. This imprecision adds to the variability in our estimate of the mean.

   * Degrees of Freedom: The concept of degrees of freedom comes into play. When estimating the sample standard deviation, we use nâˆ’1 (where n is the sample size) as the divisor instead of n to account for the fact that we're using the sample mean to estimate the population mean. This adjustment, called Bessel's correction, acknowledges the additional uncertainty in small samples.

Visualising the Normal Distribution vs t-distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# For reproducibility
np.random.seed(42)
```

```python
# Population parameters
population_mean = 50
population_std = 10

# Sample size
population_size = 10000

# Generate a large sample from the population
population_data = np.random.normal(population_mean, population_std, population_size)
```

```python
# Plot the histogram
plt.hist(population_data, bins=30, density=True, alpha=0.6, label='Population')

# Plot the probability density function (PDF) of the normal distribution
x = np.linspace(population_mean - 4*population_std, population_mean + 4*population_std, 100)
plt.plot(x, np.exp(-(x - population_mean)**2 / (2 * population_std**2)) / (population_std * np.sqrt(2 * np.pi)),
         color='red', label='Normal PDF')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```

```python
# Parameters for sampling
sample_size = 10
num_samples = 10000

# Take multiple small samples and calculate their means
# Take multiple small samples and calculate their means
sample_means = [np.mean(np.random.choice(population_data, size=sample_size)) for _ in range(num_samples)]
```

```python
# Sample sizes to consider
sample_sizes = [10, 20, 30, 1000]

# Create subplots
fig, axes = plt.subplots(len(sample_sizes), 1, figsize=(4, 10))

for i, n in enumerate(sample_sizes):
    # Generate samples
    samples = np.random.choice(population_data, size=(n, 1000))

    # Calculate sample means and standard errors
    sample_means = np.mean(samples, axis=1)
    sample_stds = np.std(samples, axis=1, ddof=1)
    standard_errors = sample_stds / np.sqrt(n)

    # Calculate t-values
    t_values = (sample_means - population_mean) / standard_errors

    # Plot the t-distribution
    x = np.linspace(-4, 4, 100)
    t_dist = t.pdf(x, df=n-1)
    axes[i].plot(x, t_dist, label=f'T Distribution (df={n-1})')

    # Plot the histogram of t-values
    axes[i].hist(t_values, bins=30, density=True, alpha=0.6)

    axes[i].set_xlabel('T-value')
    axes[i].set_ylabel('Density')
    axes[i].set_title(f'T Distribution for n={n}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

**As we see that in the above t_distribution, the tails are heaviour than a normal distribution, otherwise the distribution is still bell shaped curve and symmetrical. and as The Number of Samples Approaches the Population, the closer it gets to the Normal Distribution**
