## 1. Introduction to Uniform Distribution


### 1.1 Definition:
* A Uniform Distribution is a probability distribution where all outcomes within a given range are equally likely.

### 1.2 Characteristics:
* Constant Probability Density Function (PDF)
* Two Parameters:
  * `a` (Lower Bound)
  * `b` (Upper Bound)
* Symmetrical Shape
* **Mean:**
$$
\text{Mean} = \frac{a + b}{2}
$$

* **Variance:**
$$
\text{Variance} = \frac{(b - a)^2}{12}
$$

### 1.3 Visual Representation:
* A Uniform Distribution is typically represented by a rectangle.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(9)
a = 0 # Lower Bound
b = 100 # Upper Bound
size=1000


x = np.random.uniform(low=a, high=b, size=size)
pdf = np.ones_like(x) * 1/(b-a)

cdf = np.linspace(0, 1, size)
x_sorted = np.sort(x)

fig, (ax0, ax1) = plt.subplots(nrows=2)
ax0.hist(x, bins=30, density=True, alpha=0.6, color='g')
ax0.plot(np.linspace(a, b, size), pdf, 'r--', label="PDF")
ax1.plot(x_sorted, cdf, 'b-', label="CDF")

ax0.set_xlabel("X Values")
ax0.set_ylabel("Density")
ax0.set_title("Histogram and PDF of Uniform DIstribution")
ax0.legend()

ax1.set_xlabel("Sorted X Values")
ax1.set_ylabel("Cumulative Density")
ax1.set_title("CDF of Uniform DIstribution")
ax1.legend()

plt.tight_layout()
```

* The x-axis represents the range of possible values, and the y-axis represents the probability density.
* The Height of the rectangle is constant across the entire range.

## 2 Mathematical Formulation

#### 1. Probability Density Function (PDF)

   * A Continuous Random Variable X is said to follow a uniform distribution over the interval [a, b] if its probability density (PD) is same for all possible X population.
   * There are b - a possible elements in the distribution.
   * To get same probability for every population of X:
      * **Probability (for every \(X\)):** $$ \text{P(X)} = \frac{1}{b - a} $$
   * Hence Probability Density Function of Uniform Distribution is:
      * $$
f(x) =
\begin{cases} 
\frac{1}{b - a} & \text{for } a \leq x \leq b \\
0 & \text{otherwise}
\end{cases}
$$
         * where 
            * `a`: Lower Bound of the Distribution
            * `b`: Upper Bound of the Distribution
            * PDF is 0 for elements not in [a, b] Interval

#### 2. Cumulative Distribution Function (CDF)

The CDF of a uniform distribution is the probability that the random variable \( X \) is less than or equal to \( x \). It is given by:

$$
F(x) =
\begin{cases}
0 & \text{for } x < a \\
\frac{x - a}{b - a} & \text{for } a \leq x \leq b \\
1 & \text{for } x > b
\end{cases}
$$
#### 3. Expected Value and Variance

- **Expected Value (Mean)**:

  The expected value \( E[X] \) for a uniform distribution is:

  $$
  E[X] = \frac{a + b}{2}
  $$

- **Variance**:

  The variance $ \text{Var}(X) $ for a uniform distribution is:

  $$
  \text{Var}(X) = \frac{(b - a)^2}{12}
  $$

#### 4. Derivation of Mean and Variance

- **Mean (Expected Value)**:

  To derive the mean, we use the definition of the expected value:

  $$
  E[X] = \int_{a}^{b} x \cdot f(x) \, dx
  $$

  Substituting the PDF $ f(x) = \frac{1}{b - a} $:

  $$
  E[X] = \int_{a}^{b} x \cdot \frac{1}{b - a} \, dx
  $$

  This integrates to:

  $$
  E[X] = \frac{1}{b - a} \left[ \frac{x^2}{2} \right]_{a}^{b} = \frac{b^2 - a^2}{2(b - a)} = \frac{a + b}{2}
  $$


    * Integral of $x⋅f(x)$: The expected value (mean) of a continuous distribution is calculated using an integral. This means we are adding up all possible values of xx, weighted by their probability density $f(x)$.
    * Intuition: This gives us the average value where the distribution is centered. For a uniform distribution, the mean is simply the midpoint of the interval because the probability is evenly spread out.

- **Variance**:

  The variance is derived from:

  $$
  \text{Var}(X) = E[X^2] - (E[X])^2
  $$

  First, compute \( E[X^2] \):

  $$
  E[X^2] = \int_{a}^{b} x^2 \cdot f(x) \, dx = \frac{1}{b - a} \int_{a}^{b} x^2 \, dx
  $$

  This integrates to:

  $$
  E[X^2] = \frac{1}{b - a} \left[ \frac{x^3}{3} \right]_{a}^{b} = \frac{b^3 - a^3}{3(b - a)}
  $$

  Now compute the variance:

  $$
  \text{Var}(X) = \frac{b^3 - a^3}{3(b - a)} - \left(\frac{a + b}{2}\right)^2
  $$

  Simplifying this:

  $$
  \text{Var}(X) = \frac{(b - a)^2}{12}
  $$

    * Integral of $x^2⋅f(x)$: To find variance, we first compute the expected value of x2x2 (how spread out the squared values are) and then subtract the square of the expected value of xx (the average value).
    * Intuition: Variance tells us how much the values of xx spread out from the mean. In a uniform distribution, this spread is constant and predictable based on the range.
#### 5. Moments of the Uniform Distribution

The \( k \)-th moment of the uniform distribution about the origin is:

$$
E[X^k] = \frac{1}{b - a} \int_{a}^{b} x^k \, dx
$$

This integrates to:

$$
E[X^k] = \frac{b^{k+1} - a^{k+1}}{(k+1)(b - a)}
$$

- **First Moment (Mean)**: $ E[X] = \frac{a + b}{2} $
  
- **Second Moment**: $ E[X^2] = \frac{b^3 - a^3}{3(b - a)} $
  
- **General \( k \)-th Moment**: $ E[X^k] = \frac{b^{k+1} - a^{k+1}}{(k+1)(b - a)} $
